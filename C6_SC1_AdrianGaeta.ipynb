**13. Todo parece indicar que las variables categóricas que quedan sí son importantes para obtener la variable de salida. Convierte todas las variables categóricas que quedan a numéricas:**

a. BusinessTravel

b. Department

c. EducationField

d. Gender

e. JobRole

f. MaritalStatus

g. Attrition

```python
# Convertir las variables en categóricas
EmpleadosAttrition['BusinessTravel'] = pd.factorize(EmpleadosAttrition['BusinessTravel'].astype('category'))[0]
EmpleadosAttrition['Department'] = pd.factorize(EmpleadosAttrition['Department'].astype('category'))[0]
EmpleadosAttrition['EducationField'] = pd.factorize(EmpleadosAttrition['EducationField'].astype('category'))[0]
EmpleadosAttrition['Gender'] = pd.factorize((EmpleadosAttrition['Gender'].astype('category')))[0]
EmpleadosAttrition['JobRole'] = pd.factorize(EmpleadosAttrition['JobRole'].astype('category'))[0]
EmpleadosAttrition['MaritalStatus'] = pd.factorize(EmpleadosAttrition['MaritalStatus'].astype('category'))[0]
EmpleadosAttrition['OverTime'] = pd.factorize(EmpleadosAttrition['OverTime'].astype('category'))[0]
EmpleadosAttrition['Attrition'] = pd.factorize(EmpleadosAttrition['Attrition'].astype('category'))[0]

# Mostrar la nueva configuración del data frame
EmpleadosAttrition.info()
```

**14. Ahora debes hacer la evaluación de las variables para quedarte con las mejores. Calcula la correlación lineal de cada una de las variables con respecto al Attrition.**

```python
corrAttrition = EmpleadosAttrition.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
display(corrAttrition['Attrition'])
```

**15. Selecciona solo aquellas variables que tengan una correlación mayor o igual a 0.1, dejándolas en otro frame llamado EmpleadosAttritionFinal. No olvides mantener la variable de salidaAttrition; esto es equivalente a borrar las que no cumplen con el límite.**

```python
salidaAttrition = pd.DataFrame(corrAttrition['Attrition'][corrAttrition['Attrition'].abs() >= 0.1])
EmpleadosAttritionFinal = EmpleadosAttrition[salidaAttrition.index.to_list()]

display(salidaAttrition)
display(EmpleadosAttritionFinal)
```

**16. Crea una nueva variable llamada EmpleadosAttritionPCA formada por los componentes principales del frame EmpleadosAttritionFinal. Recuerda que el resultado del proceso PCA es un numpy array, por lo que, para hacer referencia a una columna, por ejemplo, la 0, puedes usar la instrucción EmpleadosAttritionPCA[:,0]).**

```python
pca_process = PCA(n_components=2)
EmpleadosAttritionPCA = pca_process.fit_transform(EmpleadosAttrition)
display(EmpleadosAttritionPCA)
```

**17. Agrega el mínimo número de Componentes Principales en columnas del frame EmpleadosAttritionPCA que logren explicar el 80% de la varianza, al frame EmpleadosAttritionFinal. Puedes usar la instrucción assign, columna por columna, llamando a cada unaC0, C1, etc., hasta las que vayas a agregar.**

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(EmpleadosAttritionFinal)
acc_var = np.cumsum(pca_process.explained_variance_ratio_)
components_number = int(np.searchsorted(acc_var, 0.8) + 1)

pca_process = PCA(n_components=components_number)
EmpleadosAttritionPCA = pca_process.fit_transform(x_scaled)

# Creamos una copia del dataframe para evitar warnings de modificación de vistas
EmpleadosAttritionFinal = EmpleadosAttritionFinal.copy()
EmpleadosAttritionFinal[[f'C{i}' for i in range(EmpleadosAttritionPCA.shape[1])]] = EmpleadosAttritionPCA

display(EmpleadosAttritionFinal.head())

```
